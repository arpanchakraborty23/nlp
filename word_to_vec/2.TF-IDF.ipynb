```python
import os 
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
os.chdir('../')
data=pd.read_csv('Education.csv')
data.head()
```

```python
import re
corpus=[]
for i in range(len(data)):
    txt=re.sub('[^a-zA-z]',' ',data['Text'][i])
    # tokenize
    txt=word_tokenize(txt)
    txt=[WordNetLemmatizer().lemmatize(word) for word in txt if word not in set(stopwords.words('english'))]
    # join word
    txt=' '.join(txt)
    corpus.append(txt)
    print(txt)
```

## TF-IDF

```python

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=230)
x=tfidf.fit_transform(corpus)
x=x.toarray()
```
